When preparing your data for an ANN, focus on scaling continuous numerical features while keeping one-hot encoded categorical features unchanged since they are already represented appropriately for the model.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pickle

# Load the processed dataset
train_split = pd.read_csv('train_split_processed.csv')

# Numerical features to be scaled
numerical_features = ['Age', 'Work_Experience', 'Family_Size', 'Spending_Score']

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the numerical features
train_split[numerical_features] = scaler.fit_transform(train_split[numerical_features])

# Save the scaler using pickle
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Export the scaled dataset to a new CSV file
train_split.to_csv('train_split_scaled.csv', index=False)
```
