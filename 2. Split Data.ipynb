# Get Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('df_without_duplicates.csv')
```

```python
df.shape
```

# Split Data

Note: In our case, we will implement an ANN for learning purposes to create examples of different tasks in Git repository.

For an 7951-row dataset, an 80/20 split is generally recommended for training an ANN. This split provides a substantial amount of data for training while still maintaining a reasonable size for the test set to evaluate performance.

```python
train_split, test_split = train_test_split(df, test_size=0.2, random_state=42)

print("Training set shape:", train_split.shape)
print("Testing set shape:", test_split.shape)

# Export the datasets to CSV files
train_split.to_csv('train_split.csv', index=False)
test_split.to_csv('test_split.csv', index=False)
```

# Train and Test Set Data Distribution Analysis

```python
# Plot count plot for the target variable 'Segmentation'
fig, axes = plt.subplots(1, 2, figsize=(20, 6))
sns.countplot(data=train_split, x='Segmentation', ax=axes[0])
axes[0].set_title('Train Set: Distribution of Segmentation')
sns.countplot(data=test_split, x='Segmentation', ax=axes[1])
axes[1].set_title('Test Set: Distribution of Segmentation')
plt.show()

# List of categorical columns
categorical_columns = ['Gender', 'Ever_Married', 'Graduated', 'Profession', 'Spending_Score', 'Var_1']

# Plot count plots for each categorical variable
for column in categorical_columns:
    fig, axes = plt.subplots(1, 2, figsize=(20, 6))
    sns.countplot(data=train_split, x=column, ax=axes[0])
    axes[0].set_title(f'Train Set: Distribution of {column}')
    sns.countplot(data=test_split, x=column, ax=axes[1])
    axes[1].set_title(f'Test Set: Distribution of {column}')
    plt.show()

# List of numerical columns
numerical_columns = ['Age', 'Work_Experience', 'Family_Size']

# Plot histograms for each numerical variable
for column in numerical_columns:
    fig, axes = plt.subplots(1, 2, figsize=(20, 6))
    sns.histplot(data=train_split, x=column, kde=True, ax=axes[0])
    axes[0].set_title(f'Train Set: Distribution of {column}')
    sns.histplot(data=test_split, x=column, kde=True, ax=axes[1])
    axes[1].set_title(f'Test Set: Distribution of {column}')
    plt.show()

# Plot box plots for each numerical variable
for column in numerical_columns:
    fig, axes = plt.subplots(1, 2, figsize=(20, 6))
    sns.boxplot(data=train_split, x=column, ax=axes[0])
    axes[0].set_title(f'Train Set: Box Plot of {column}')
    sns.boxplot(data=test_split, x=column, ax=axes[1])
    axes[1].set_title(f'Test Set: Box Plot of {column}')
    plt.show()
```

Analyzing plots we can see that both sets are representative of the complete dataset.
